```python
def categorize_experience(years):
    if years < 5:
        return "< 5 Years"
    elif years < 10:
        return "5-10 Years"
    elif years < 15:
        return "10-15 Years"
    elif years < 20:
        return "15-20 Years"
    elif years < 25:
        return "20-25 Years"
    else:
        return "25+ Years"

df["Experience_Category"] = df.iloc[:, 0].apply(categorize_experience)

df.head()
```

```python
grouped_data = df.groupby("Experience_Category")["compensation"].apply(list)

# separating list by assigning category
lessthan5 = grouped_data.get("< 5 Years") if "< 5 Years" in grouped_data else []
fiveto10 = grouped_data.get("5-10 Years") if "5-10 Years" in grouped_data else []
tento15 = grouped_data.get("10-15 Years") if "10-15 Years" in grouped_data else []
fifteento20 = grouped_data.get("15-20 Years") if "15-20 Years" in grouped_data else []
twentyto25 = grouped_data.get("20-25 Years") if "20-25 Years" in grouped_data else []
moretha25 = grouped_data.get("25+ Years") if "25+ Years" in grouped_data else []
```

```python
df = df.copy()

df.iloc[:, 0] = df.iloc[:, 0].astype(float)
df.iloc[:, 5] = df.iloc[:, 5].astype(float)

df.iloc[:, 1] = df.iloc[:, 1].astype(str) == "TRUE"
df.iloc[:, 2] = df.iloc[:, 2].astype(str) == "TRUE"
df.iloc[:, 3] = df.iloc[:, 3].astype(str) == "TRUE"

df.iloc[:, 4] = df.iloc[:, 4].replace("None", None)

df.info()
df.head()
```
