```python

import time

name = input("What's your name? ")

print("\n Hey " + name)
time.sleep(0.5)

print("\n Welcome to the Game World.")
time.sleep(0.5)

print("Time to play ! ")
time.sleep(1)



print( "\n Start guessing...")
time.sleep(0.5)

word = "python"

guesses = ''

turns = 15

while turns > 0:   

    failed = 0             


    for character in word:
        if character in guesses:
            print(character),
            
        else:
            print("_"),
            failed += 1
        
    if failed == 0:
        print( "\n\t Congratulations ! \n\t You won. \n\t The word was python "  )
        
        break
        
    print
    
    guess = input("guess a character:") 
    
    guesses += guess         

    if guess not in word:  
    
        turns -= 1      
        
        print( "Wrong." )
        
        print( "You have ", + turns, ' more guesses.') 
        
        if turns == 0:        
        
            print( "\n\t You Lose ! \n\t Better Luck Next Time. "  )
        
```
